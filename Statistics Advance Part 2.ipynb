#### Q. 1 What is hypothesis testing in statistics

- Hypothesis testing is a statistical method used to make inferences about a population based on sample data.

- Hypothesis testing is a formal statistical procedure where:
- We formulate two competing hypotheses about a population parameter:
- The null hypothesis (Hâ‚€) â€” the default assumption.
- The alternative hypothesis (Hâ‚ or Ha) â€” the claim we want to test.
- We collect sample data and perform a statistical test.
- Based on the result (e.g. p-value or test statistic), we decide whether to:
- Reject Hâ‚€ (evidence supports the claim), or
- Fail to reject Hâ‚€ (not enough evidence to support the claim).

#### Q.2 What is the null hypothesis, and how does it differ from the alternative hypothesis

- The null hypothesis (Hâ‚€) is the default or status quo assumption. We assume it to be true unless there is strong statistical evidence against it.
- We only reject Hâ‚€ if the sample data provides enough evidence that supports an alternative claim.

- The alternative hypothesis (Hâ‚ or Ha) is the claim we want to test. It's usually the opposite of Hâ‚€, and it represents a new effect, difference, or change from the norm.
- We use sample data to test whether there's enough statistical evidence to support the alternative hypothesis.

#### Q.3 What is the significance level in hypothesis testing, and why is it important?

- The significance level, denoted by Î± (alpha), is the threshold probability for deciding whether to reject the null hypothesis (Hâ‚€). It represents the maximum risk of making a Type I error â€” that is, rejecting a true null hypothesis.
- If Î± = 0.05, It means :
- â€œI am willing to accept a 5% chance of wrongly rejecting the null hypothesis.â€
- Example:

- You're testing whether a new drug is more effective than the existing one.

- Hâ‚€: The new drug is not more effective.
- Hâ‚: The new drug is more effective.
- Î± = 0.05: You're saying â€œIâ€™ll reject Hâ‚€ (approve the new drug) only if the chance of me being wrong is less than 5%."
  

#### Q.4 What does a P-value represent in hypothesis testing

- The p-value is the probability of getting the observed results, or more extreme results, assuming the null hypothesis (Hâ‚€) is true.
- If Hâ‚€ is true, the p-value tells you:
- â€œWhat are the chances that I would see this result (or something more extreme) just by random chance?â€

#### Q.6 What are Type 1 and Type 2 errors in hypothesis testing

- **Type I Error (False Positive)**
- You reject a null hypothesis that is actually true.
- You conclude there's an effect when there's not.
- Controlled by the significance level (Î±) (commonly 0.05)
-m **Example:**
- You create a new drug to lower blood pressure.
- Hâ‚€: The new drug has no effect.
- Hâ‚: The new drug does work.
- You perform a test and get p = 0.03, so you reject Hâ‚€.
- But in reality, the drug actually does nothing.
- You made a Type I error â€” falsely claiming the drug works.
- Think of it as: ---"You believed a lie."


- **Type II Error (False Negative)**
- You fail to reject a null hypothesis that is actually false.
- You miss a real effect.
- Controlled by power of the test (1 - Î²)
- **Example:*8
- Same drug test.
- Hâ‚€: The drug has no effect.
- Hâ‚: The drug does work.
- You get p = 0.08, so you fail to reject Hâ‚€.
- But in reality, the drug does work.
- You made a Type II error â€” failed to detect the effect.
-  Think of it as: -- "You missed the truth."

#### Q.7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?


- A one-tailed test checks if the sample mean is either greater than or less than the population mean â€” but only in one specific direction.
- A two-tailed test checks if the sample mean is different from the population mean â€” in either direction, whether it's higher or lower.
- In a one-tailed test, the entire rejection region (based on the significance level) is on one side of the distribution.
- In a two-tailed test, the rejection region is split between both sides of the distribution (half on the left, half on the right).
- A one-tailed test is used when your hypothesis is directional (e.g., "greater than" or "less than").
- A two-tailed test is used when your hypothesis is non-directional (e.g., "not equal to").
- Two-tailed tests are more conservative because they require stronger evidence to reject the null hypothesis.

#### Q.3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (should be 0 or 1)
    p (float): The probability of success (0 â‰¤ p â‰¤ 1)

    Returns:
    float: The probability P(X = x)
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0  # Bernoulli distribution only defined for x = 0 or 1


print("P(X = 0):", bernoulli_pdf(0, 0.5))
print("P(X = 1):", bernoulli_pdf(1, 0.7))
print("P(X = 2):", bernoulli_pdf(2, 0.7)) 
```

#### Q.6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt


outcomes = np.arange(1, 7)
probability = 1 / len(outcomes)


cdf = np.cumsum([probability] * len(outcomes))

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(outcomes, cdf, where='post', color='royalblue', label='CDF')
plt.scatter(outcomes, cdf, color='red', zorder=5)
plt.title("CDF of a Discrete Uniform Distribution (Fair Die)")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(outcomes)
plt.ylim(0, 1.1)
plt.legend()
plt.show()
```

#### Q.9 Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot Z-scores (line plot and histogram)
    plt.figure(figsize=(14, 5))

    # Line plot
    plt.subplot(1, 2, 1)
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-Scores (Line Plot)')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)

    # Histogram
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=10, color='skyblue', edgecolor='black')
    plt.title('Z-Scores (Histogram)')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()

    return z_scores


sample_data = np.random.normal(loc=50, scale=10, size=30)
z_scores = calculate_and_plot_z_scores(sample_data)
```

#### Q.10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

def demonstrate_clt(sample_size=30, num_samples=1000):
    '''Step 1: Generate non-normal population data (exponential distribution)'''
    population = np.random.exponential(scale=2.0, size=100000)

    ''' Step 2: Draw multiple samples and compute sample means'''
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size)
        sample_means.append(np.mean(sample))


    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.hist(population, bins=50, color='skyblue', edgecolor='black', density=True)
    plt.title('Original Population (Exponential)')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.5)

    # Step 4: Plot the distribution of sample means
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='orange', edgecolor='black', density=True)
    plt.title(f'Distribution of Sample Means\n(sample size = {sample_size}, n = {num_samples})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()


demonstrate_clt(sample_size=50, num_samples=1000)
```

#### Q.14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal

#### Q.5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0=50, mu_1=52, sigma=15, n=40, alpha=0.05):
    se = sigma / np.sqrt(n)

    z_critical = norm.ppf(1 - alpha / 2)
    lower_crit = mu_0 - z_critical * se
    upper_crit = mu_0 + z_critical * se
    beta = norm.cdf((upper_crit - mu_1) / se) - norm.cdf((lower_crit - mu_1) / se)
    power = 1 - beta

    print(f"Z-critical: Â±{z_critical:.3f}")
    print(f"Type I Error (Î±): {alpha}")
    print(f"Type II Error (Î²): {beta:.4f}")
    print(f"Power of the test (1 - Î²): {power:.4f}")

    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    h0_pdf = norm.pdf(x, mu_0, se)
    h1_pdf = norm.pdf(x, mu_1, se)

    plt.figure(figsize=(12, 6))
    plt.plot(x, h0_pdf, label="Hâ‚€ distribution", color="blue")
    plt.plot(x, h1_pdf, label="Hâ‚ distribution", color="green")


    plt.fill_between(x, 0, h0_pdf, where=(x < lower_crit), color='red', alpha=0.4, label="Type I Error (Î±)")
    plt.fill_between(x, 0, h0_pdf, where=(x > upper_crit), color='red', alpha=0.4)
    plt.fill_between(x, 0, h1_pdf, where=(x > lower_crit) & (x < upper_crit), color='orange', alpha=0.4, label="Type II Error (Î²)")
    plt.axvline(lower_crit, color='black', linestyle='--', label=f'Critical values: {lower_crit:.2f}, {upper_crit:.2f}')
    plt.axvline(upper_crit, color='black', linestyle='--')

    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()
```

#### Q.11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(prior_D, prob_positive_given_D, prob_positive_given_not_D):

    prior_not_D = 1 - prior_D  


    marginal_positive = (prob_positive_given_D * prior_D) + \
                        (prob_positive_given_not_D * prior_not_D)


    posterior = (prob_positive_given_D * prior_D) / marginal_positive
    return posterior


prior_D = 0.01  # 1% have the disease
prob_positive_given_D = 0.99  # Test sensitivity
prob_positive_given_not_D = 0.05  # False positive rate


posterior = bayes_theorem(prior_D, prob_positive_given_D, prob_positive_given_not_D)
print(f"Probability of having disease given a positive test result: {posterior:.4f}")
```

```python
import numpy as np
import scipy.stats as stat

# Observed and Expected Frequencies
expected_data = [8, 6, 7, 9, 6, 9, 7]
observed_data = [7, 8, 6, 9, 9, 6, 7]


chisquare_stats, p_value = stat.chisquare(f_obs=observed_data, f_exp=expected_data)


significance = 0.05
dof = len(observed_data) - 1
chi_sq_critical = stat.chi2.ppf(1 - significance, dof)


print(f"Chi-square Statistic: {chisquare_stats:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Chi-square Critical Value: {chi_sq_critical:.4f}")


if chisquare_stats > chi_sq_critical:
    print("Reject the null hypothesis: The data does NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The data fits the expected distribution.")
```
