#### Q.3 What is the difference between discrete and continuous distributions?

 **Discrete Distribution:**
Deals with discrete random variables (countable values).

Takes values like 0, 1, 2, 3, etc. ‚Äî no decimals between them.

Uses a Probability Mass Function (PMF) to assign probabilities.

Can assign a positive probability to a specific value (e.g., P(X = 2) = 0.3).

The total sum of all probabilities is always 1.

Represented visually using bar graphs.

Example: Number of goals scored in a soccer match.

**Continuous Distribution:**
Deals with continuous random variables (uncountably infinite values).

Takes any value within an interval (e.g., 2.35, 2.351, 2.3517, etc.).

Uses a Probability Density Function (PDF) to describe probabilities.

The probability of any exact value is always 0 (e.g., P(X = 2.5) = 0).

Probability is found by calculating the area under the curve over an interval.

The total area under the curve is always 1.

Represented visually using smooth curves.

Example: Time taken to complete a task or height of individuals.

#### Q.5 How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

- On the other hand, the Probability Distribution Function (PDF) ‚Äî which is a general term ‚Äî represents how the probabilities are spread across values of the random variable. #### Q.13 What is the Central Limit Theorem (CLT), and why is it critical in statistics?

- Even if the original data is skewed, bimodal, or non-normal, we can still use normal-based methods when working with sample means.
- Foundation for inferential statistics:
- CLT is the backbone of many statistical techniques like:
- Confidence intervals
- Hypothesis testing
- Z-tests and t-tests
- It simplifies complex problems:
- It turns unknown or complex distributions into something predictable and manageable‚Äîthe normal distribution.

#### Q.3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (should be 0 or 1)
    p (float): The probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
    float: The probability P(X = x)
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0  # Bernoulli distribution only defined for x = 0 or 1


print("P(X = 0):", bernoulli_pdf(0, 0.7))
print("P(X = 1):", bernoulli_pdf(1, 0.7))
print("P(X = 2):", bernoulli_pdf(2, 0.7)) 
```

#### Q.6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

```python
import numpy as np
import matplotlib.pyplot as plt


outcomes = np.arange(1, 7)
probability = 1 / len(outcomes)


cdf = np.cumsum([probability] * len(outcomes))

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(outcomes, cdf, where='post', color='royalblue', label='CDF')
plt.scatter(outcomes, cdf, color='red', zorder=5)
plt.title("CDF of a Discrete Uniform Distribution (Fair Die)")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(outcomes)
plt.ylim(0, 1.1)
plt.legend()
plt.show()
```

#### Q.9 Write a Python function to calculate Z-scores from a dataset and plot them

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot Z-scores (line plot and histogram)
    plt.figure(figsize=(14, 5))

    # Line plot
    plt.subplot(1, 2, 1)
    plt.plot(z_scores, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-Scores (Line Plot)')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)

    # Histogram
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=10, color='skyblue', edgecolor='black')
    plt.title('Z-Scores (Histogram)')
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()

    return z_scores


sample_data = np.random.normal(loc=50, scale=10, size=100)
z_scores = calculate_and_plot_z_scores(sample_data)
```

#### Q.10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

def demonstrate_clt(sample_size=30, num_samples=1000):
    '''Step 1: Generate non-normal population data (exponential distribution)'''
    population = np.random.exponential(scale=2.0, size=100000)

    ''' Step 2: Draw multiple samples and compute sample means'''
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size)
        sample_means.append(np.mean(sample))


    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.hist(population, bins=50, color='skyblue', edgecolor='black', density=True)
    plt.title('Original Population (Exponential)')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.5)

    # Step 4: Plot the distribution of sample means
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='orange', edgecolor='black', density=True)
    plt.title(f'Distribution of Sample Means\n(sample size = {sample_size}, n = {num_samples})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.5)

    plt.tight_layout()
    plt.show()


demonstrate_clt(sample_size=50, num_samples=1000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mu, sigma):
    """Calculate the Z-score for a given data point."""
    return (x - mu) / sigma

def plot_standard_normal_with_z(x, mu, sigma):
    """Plot standard normal distribution and mark the Z-score."""
    z = calculate_z_score(x, mu, sigma)
    
    # Generate x values for the standard normal curve
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label="Standard Normal Distribution", color="blue")
    plt.axvline(z, color='red', linestyle='--', label=f"Z-score = {z:.2f}")
    plt.fill_between(x_vals, y_vals, where=(x_vals <= z), color='red', alpha=0.2)
    
    plt.title("Z-score Compared to Standard Normal Distribution")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()
    
    print(f"Z-score for x = {x} (mean = {mu}, std = {sigma}): {z:.2f}")


calculate_z_score(75,70,5)
plot_standard_normal_with_z(75,70,5)
```

#### Q.15 Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    
    # Standard error
    standard_error = population_std / np.sqrt(sample_size)
    
    # Z-statistic
    z = (sample_mean - population_mean) / standard_error
    
    # Critical Z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)
    

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-statistic = {z:.2f}")
    print(f"Critical Z-value = ¬±{z_critical:.2f}")
    

    if abs(z) > z_critical:
        print("Reject the null hypothesis (significant difference).")
    else:
        print("Fail to reject the null hypothesis (no significant difference).")


sample = [172, 174, 169, 171, 173, 175, 170, 168, 176, 169]


z_test(sample_data=sample, population_mean=170, population_std=3)
```

#### Q.16 Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval = ({lower_bound:.2f}, {upper_bound:.2f})")

    return (lower_bound, upper_bound)


data = [45, 47, 49, 46, 48, 47, 50, 49, 46, 48]


confidence_interval(data)
```

#### Q.17 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt


np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50


sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

'''Calculate standard error and margin of error'''
confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
standard_error = sample_std / np.sqrt(n)
margin_of_error = t_critical * standard_error


ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error


print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence * 100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")


plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='--', label='Lower Bound')
plt.axvline(ci_upper, color='green', linestyle='--', label='Upper Bound')
plt.title('Histogram of Sample Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_stats(size=1000, mean=0, std=1):

    data = np.random.normal(loc=mean, scale=std, size=size)


    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  


    print(f"Sample Size: {size}")
    print(f"Mean: {np.mean(data):.4f}, Std Dev: {np.std(data):.4f}")
    print(f"Skewness: {data_skewness:.4f} (expected ‚âà 0 for normal)")
    print(f"Kurtosis: {data_kurtosis:.4f} (expected ‚âà 0 for normal)")

    return data_skewness, data_kurtosis


analyze_distribution_stats()
```
